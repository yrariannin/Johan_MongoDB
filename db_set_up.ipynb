## Use JSONSCHEMA instead of json.dump

```python
specimen={
    "type":"object",
    "required": [
    "_id",
    "company_name",
    "company_site",
    "disassembly_day",
    "installation_date",
    "nominal_diametre",
    "nominal_thickness",
    "receipt_day",
    "specimen _id",
    "specimen_name"
  ],
  "properties": {
    "_id": {
      "$ref": "#/$defs/ObjectId"
    },
    "company_name": {
      "type": "string"
    },
    "company_site": {
      "type": "string"
    },
    "disassembly_day": {
      "$ref": "#/$defs/Date"
    },
    "installation_date": {
      "$ref": "#/$defs/Date"
    },
    "nominal_diametre": {
      "$ref": "#/$defs/Double"
    },
    "nominal_thickness": {
      "$ref": "#/$defs/Double"
    },
    "receipt_day": {
      "$ref": "#/$defs/Date"
    },
    "specimen _id": {
      "type": "string"
    },
    "specimen_name": {
      "type": "string"
    }
  },
  "$defs": {
    "ObjectId": {
      "type": "object",
      "properties": {
        "$oid": {
          "type": "string",
          "pattern": "^[0-9a-fA-F]{24}$"
        }
      },
      "required": [
        "$oid"
      ],
      "additionalProperties": false
    },
    "Double": {
      "oneOf": [
        {
          "type": "number","exclusiveMinimum": 0
        },
        {
          "type": "object",
          "properties": {
            "$numberDouble": {
              "enum": [
                "Infinity",
                "-Infinity",
                "NaN"
              ]
            }
          }
        }
      ]
    },
    "Date": {
      "type": "object",
      "properties": {
        "$date": {
          "type": "string",
          "format": "date"
        }
      },
      "required": [
        "$date"
      ],
      "additionalProperties": false
    }
  }
}
```
